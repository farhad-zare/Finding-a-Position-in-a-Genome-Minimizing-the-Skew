```python
f=open('rosalind_ba1f.txt','r')
```

```python
data = f.read()
f.close()
```

```python
data
```

```python
dic={'C':-1,'G':1}
```

```python
L=[0 for i in range(0,len(data)+1)]
for i in range(0,len(data)):
    if data[i] in dic:
        L[i+1]=L[i]+dic[data[i]]
    else:
        L[i+1]=L[i]
```

```python
m=min(L)
T=True
while T:
    if min(L)==m:
        print(L.index(min(L)))
        L[L.index(min(L))]=max(L)
    else:
        T=False
```
